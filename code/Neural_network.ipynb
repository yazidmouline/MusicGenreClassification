```python
data = pd.read_csv('Music_data_set.csv').drop(['Unnamed: 0', 'song'], axis=1)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data[['mean_mfccs', 'mean_chroma_stft', 'tempo', 'pulse', 'contrast', 'zero_crossing']] = sc.fit_transform(data[['mean_mfccs', 'mean_chroma_stft', 'tempo', 'pulse', 'contrast', 'zero_crossing']])
```

```python
y = data['genre']
X = data.drop(['genre'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size= 0.25, random_state=27)
```

```python
x_train
```

```python
clf = MLPClassifier(hidden_layer_sizes=(7,10,10), max_iter=1000, alpha=0.0001,activation = 'relu',
                    solver='sgd', verbose=10,  random_state=1,tol=0.000000001)
```

```python
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, center=True)
plt.show()
```
